シグモイド関数
    \begin{equation*}
      \sigma(x) = \frac{1}{1+e^{-x}}
    \end{equation*}
  の増減，極値，凹凸を調べ，曲線$y=\sigma(x)$の概形を描け．

      \sigma(x)(1-\sigma(x))

````python
y = 1/(1+exp(-x))
y
````

````python
dy = diff(y, x)
dy
````

````python
%matplotlib inline

plot(y,y*(1-y)+0.01,dy,(x,-10,10))
````

# 3 センター試験原題(20点)
(2019大学入試センター試験　数学II・B 第2問(1),(2))

$p, q$ を実数とし，
関数$f(x)=x^3 + p x^2 +qx$ は$x=-1$で極値2を取るとする．
また，座標平面上の曲線$y=f(x)$を$C$,放物線$y=-kx^2$ を$D$,
放物線$D$上の点$(a, -ka^2)$をAとする．
ただし，　$k \gt 0, a\gt0$である．

(1) 関数$f(x)$が$x=-1$で極値をとるので，
$f'(-1) = \fbox{ ア }$である．
これと$f(-1)=2$より，　$p=\fbox{ イ }\,, q={ \fbox{ ウエ }}$である．
よって$f(x)$は$x= \fbox{ オ }$で極小値$ \fbox{ カキ }$をとる．


````python
from sympy import *
init_session()

a,k,p,q,x,y=symbols('a k p q x y')
````

````python
fx = x**3+p*x**2+q*x
fx
````

````python
df = diff(fx, x)
df
````

````python
x0=-1
y0=2
````

````python
e1=fx.subs({x:x0})-y0
e2=df.subs({x:x0})
````

````python
{e1,e2}
````

````python
subs_pq=solve({e1,e2},{p,q})
subs_pq
````

````python
sol1 = solve(df.subs(subs_pq),x)
sol1
````

````python
fx.subs(subs_pq).subs({x:sol1[1]})
````

````python
fx0 = fx.subs(subs_pq)
fx0
````

````python
%matplotlib inline

plot(fx0,-x**2,(x,-2,2))
````

(2) 点Aにおける放物線$D$の接線を$l$とする．
$D$と$l$および$x$軸で囲まれた図形の面積$S$を
$a$と$k$を用いて表そう．

$l$の方程式は
\begin{equation*}
  y = \fbox{ クケ }\,kax + \,ka^{ \fbox{ コ }}　... (1)
\end{equation*}
と表せる．
$l$と$x$軸の交点の$x$座標は
$\frac{\fbox{ サ }}{\fbox{ シ }}$であり，
$D$と$x$軸および
直線$x=a$で囲まれた図形の面積は
$\frac{k}{\fbox{ ス }}a^{\fbox{ セ }}$である．
よって，$S=\frac{k}{\fbox{ ソタ }} a^{\fbox{ セ }}$である．

````python
fx2=-k*x**2
fx2
````

````python
aa= diff(fx2,x).subs({x:a})
aa
````

````python
ll = expand(aa*(x-a)+fx2.subs({x:a}))
ll
````

````python
x0 = solve(ll,x)
x0
````

````python
Sd = -integrate(fx2,(x,0,a))
Sd
````

````python
Sll=-integrate(ll,(x,a/2,a))
Sll
````

````python
SS=Sd - Sll
SS
````
