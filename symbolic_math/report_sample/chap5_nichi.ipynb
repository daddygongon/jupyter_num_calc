## 二乗誤差関数

```python
# リスト 5-1-(6)
from mpl_toolkits.mplot3d import Axes3D

# 平均誤差関数 ------------------------------
def mse_line(x, t, w):
    y = w[0] * x + w[1]
    mse = np.mean((y - t)**2)
    return mse

# 計算 --------------------------------------
xn = 100  # 等高線表示の解像度
w0_range = [-25, 25]
w1_range = [120, 170]
w0 = np.linspace(w0_range[0], w0_range[1], xn)
w1 = np.linspace(w1_range[0], w1_range[1], xn)
ww0, ww1 = np.meshgrid(w0, w1)
J = np.zeros((len(w0), len(w1)))
for i0 in range(len(w0)):
    for i1 in range(len(w1)):
        J[i1, i0] = mse_line(X, T, (w0[i0], w1[i1]))
        
# 表示 --------------------------------------
plt.figure(figsize=(9.5, 4))
plt.subplots_adjust(wspace=0.5)

ax = plt.subplot(1, 2, 1, projection='3d')
ax.plot_surface(ww0, ww1, J, rstride=10, cstride=10, alpha=0.3, 
                color='blue', edgecolor='black')
ax.set_xticks([-20, 0, 20])
ax.set_yticks([120, 140, 160])
ax.view_init(20, -60)

plt.subplot(1, 2, 2)
cont = plt.contour(ww0, ww1, J, 30, colors='black',
                   levels=[100, 1000, 10000, 100000])
cont.clabel(fmt='%d', fontsize=8)
plt.grid(True)
plt.show()
```

- 計算の部分で傾きのw0と切片w1に適当な数値を入れていることがわかる．
- 右のグラフは左のグラフのw1方向．窪んでいるところの右の縦方向の値がみにくいため等高線での表示．地理で習った山の高さとか表すやつ．
- 線の密集してる130-140あたりが最小値ということがわかる．

## 5.1.3パラメータを求める（勾配法）

- 勾配法とは，アルゴリズム．初期位置として決めた点から傾きを調べていって．傾きが小さくなっているところを徐々に更新して行って最終的に最も小さくなる点に到達すること．
- 教科書ではこの点をお椀の底と表現している．

```python
#そして勾配を計算する関数がdmse_line(x,t,w)
# リスト 5-1-(7)
# 平均二乗誤差の勾配 ------------------------
def dmse_line(x, t, w):
    y = w[0] * x + w[1]
    d_w0 = 2 * np.mean((y - t) * x)
    d_w1 = 2 * np.mean(y - t) 
    return d_w0, d_w1
```

```python
# リスト 5-1-(8)
#w=[10, 165]での勾配を求める．
d_w = dmse_line(X, T, [10, 165])
print(np.round(d_w, 1))
```

```python
# リスト 5-1-(9)
# 勾配法 ------------------------------------
def fit_line_num(x, t):
    w_init = [10.0, 165.0]  # 初期パラメータ
    alpha = 0.001  # 学習率
    tau_max = 100000  # 繰り返しの最大数
    eps = 0.1  # 繰り返しをやめる勾配の絶対値のしきい値
    w_hist = np.zeros([tau_max, 2])
    w_hist[0, :] = w_init
    for tau in range(1, tau_max):
        dmse = dmse_line(x, t, w_hist[tau - 1])
        w_hist[tau, 0] = w_hist[tau - 1, 0] - alpha * dmse[0]
        w_hist[tau, 1] = w_hist[tau - 1, 1] - alpha * dmse[1]
        if max(np.absolute(dmse)) < eps: # 終了判定
            break
    w0 = w_hist[tau, 0]
    w1 = w_hist[tau, 1]
    w_hist = w_hist[:tau, :]
    return w0, w1, dmse, w_hist


# メイン ------------------------------------
plt.figure(ﬁgsize=(4, 4))
# MSE の等高線表示
wn = 100  # 等高線解像度
w0_range = [-25, 25]
w1_range = [120, 170]
w0 = np.linspace(w0_range[0], w0_range[1], wn)
w1 = np.linspace(w1_range[0], w1_range[1], wn)
ww0, ww1 = np.meshgrid(w0, w1)
J = np.zeros((len(w0), len(w1)))
for i0 in range(wn):
    for i1 in range(wn):
        J[i1, i0] = mse_line(X, T, (w0[i0], w1[i1]))
cont = plt.contour(ww0, ww1, J, 30, colors='black',
                   levels=(100, 1000, 10000, 100000))
cont.clabel(fmt='%1.0f', fontsize=8)
plt.grid(True)
# 勾配法呼び出し
W0, W1, dMSE, W_history = fit_line_num(X, T)
# 結果表示
print(' 繰り返し回数 {0}'.format(W_history.shape[0]))
print('W=[{0:.6f}, {1:.6f}]'.format(W0, W1))
print('dMSE=[{0:.6f}, {1:.6f}]'.format(dMSE[0], dMSE[1]))
print('MSE={0:.6f}'.format(mse_line(X, T, [W0, W1])))
plt.plot(W_history[:, 0], W_history[:, 1], '.-',
         color='gray', markersize=10, markeredgecolor='cornflowerblue')
plt.show()
```

- fit_line_num(x,t)はデータx,tを引数にして，mse_lineを最小にする値wを返す．
- wは初期値w_init={10, 165}からdmse_lineで求めた勾配でwを更新しお椀の底に到達したら終了

```python
#それでは　求めたお椀の底が正しいのか確かめて見よう

# リスト 5-1-(10)
# 線の表示 ----------------------------------
def show_line(w):
    xb = np.linspace(X_min, X_max, 100)
    y = w[0] * xb + w[1]
    plt.plot(xb, y, color=(.5, .5, .5), linewidth=4)

# メイン ------------------------------------
plt.figure(figsize=(4, 4))
W=np.array([W0, W1])
mse = mse_line(X, T, W)
print("w0={0:.3f}, w1={1:.3f}".format(W0, W1))
print("SD={0:.3f} cm".format(np.sqrt(mse)))
show_line(W)
plt.plot(X, T, marker='o', linestyle='None', 
         color='cornflowerblue', markeredgecolor='black')
plt.xlim(X_min, X_max)
plt.grid(True)
plt.show()
```

## 直線モデルパラメータの解析解


- 勾配法は繰り返しの計算で"近似的"な解を求めるもので，このような解を"数値解"と呼ぶ．
- ただし，直線モデルの場合は近似的ではなく，方程式を解くことによって"厳密な解"を求めることができる．このような解を "解析解" と呼ぶ．
- 解析解のメリットは一回で早く最適なwが出る．詳しくはp160

```python
# リスト 5-1-(11)
# 解析解 ------------------------------------
def fit_line(x, t):
    mx = np.mean(x)
    mt = np.mean(t)
    mtx = np.mean(t * x)
    mxx = np.mean(x * x)
    w0 = (mtx - mt * mx) / (mxx - mx**2)
    w1 = mt - w0 * mx
    return np.array([w0, w1])
# メイン ------------------------------------
W = fit_line(X, T)
print("w0={0:.3f}, w1={1:.3f}".format(W[0], W[1]))
mse = mse_line(X, T, W)
print("SD={0:.3f} cm".format(np.sqrt(mse)))
plt.figure(figsize=(4, 4))
show_line(W)
plt.plot(X, T, marker='o', linestyle='None', 
         color='cornflowerblue', markeredgecolor='black')
plt.xlim(X_min, X_max)
plt.grid(True)
plt.show()
```

これが解析解

## 面モデルパラメータの解析解

```python
# リスト 5-1-(16)
# 解析解 ------------------------------------
def fit_plane(x0, x1, t):
    c_tx0 = np.mean(t * x0) - np.mean(t) * np.mean(x0)
    c_tx1 = np.mean(t * x1) - np.mean(t) * np.mean(x1)
    c_x0x1 = np.mean(x0 * x1) - np.mean(x0) * np.mean(x1)
    v_x0 = np.var(x0)
    v_x1 = np.var(x1)
    w0 = (c_tx1 * c_x0x1 - v_x1 * c_tx0) / (c_x0x1**2 - v_x0 * v_x1)
    w1 = (c_tx0 * c_x0x1 - v_x0 * c_tx1) / (c_x0x1**2 - v_x0 * v_x1)
    w2 = -w0 * np.mean(x0) - w1 * np.mean(x1) + np.mean(t)
    return np.array([w0, w1, w2])

# メイン ------------------------------------
plt.figure(figsize=(6, 5))
ax = plt.subplot(1, 1, 1, projection='3d')
W = fit_plane(X0, X1, T)
print("w0={0:.1f}, w1={1:.1f}, w2={2:.1f}".format(W[0], W[1], W[2]))
show_plane(ax, W)
show_data2(ax, X0, X1, T)
mse = mse_plane(X0, X1, T, W)
print("SD={0:.2f} cm".format(np.sqrt(mse)))
plt.show()
```

# 5.4　線形基底関数モデル

- xが一次元として．直線ではなくて，曲線の方がデータ的にはあっているかもしれない．
- だから曲線を表していくよってところ．そこで線形基底関数モデルを用いる．
- 詳細はp179．

```python
# リスト 5-2-(1)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# データのロード ----------------------------
outfile = np.load('ch5_data.npz')
X = outfile['X']
X_min = 0
X_max = outfile['X_max']
X_n = outfile['X_n']
T = outfile['T']
```

```python
# リスト 5-2-(2)
# ガウス関数 ---------------------------------
def gauss(x, mu, s):
    return np.exp(-(x - mu)**2 / (2 * s**2))
```

```python
# リスト 5-2-(3)
# メイン ------------------------------------
M = 4
plt.figure(figsize=(4, 4))
mu = np.linspace(5, 30, M)
s = mu[1] - mu[0]     # (A)
xb = np.linspace(X_min, X_max, 100)
for j in range(M):
    y = gauss(xb, mu[j], s)
    plt.plot(xb, y, color='gray', linewidth=3)
plt.grid(True)
plt.xlim(X_min, X_max)
plt.ylim(0, 1.2)
plt.show()
```

```python
# リスト 5-2-(4)
# 線形基底関数モデル ----------------
def gauss_func(w, x):
    m = len(w) - 1
    mu = np.linspace(5, 30, m)
    s = mu[1] - mu[0]
    y = np.zeros_like(x)     # x と同じサイズで要素が 0 の行列 y を作成
    for j in range(m):
        y = y + w[j] * gauss(x, mu[j], s)
    y = y + w[m]
    return y
```

```python
# リスト 5-2-(5)
# 線形基底関数モデル MSE ----------------
def mse_gauss_func(x, t, w):
    y = gauss_func(w, x)
    mse = np.mean((y - t)**2)
    return mse
```

```python
# リスト 5-2-(6)
# 線形基底関数モデル 厳密解 -----------------
def fit_gauss_func(x, t, m):
    mu = np.linspace(5, 30, m)
    s = mu[1] - mu[0]
    n = x.shape[0]
    phi = np.ones((n, m+1))
    for j in range(m):
        phi[:, j] = gauss(x, mu[j], s)
    phi_T = np.transpose(phi)
    
    b = np.linalg.inv(phi_T.dot(phi))
    c = b.dot(phi_T)
    w = c.dot(t)
    return w
```

```python
# リスト 5-2-(7)
# ガウス基底関数表示 -----------------------
def show_gauss_func(w):
    xb = np.linspace(X_min, X_max, 100)
    y = gauss_func(w, xb)
    plt.plot(xb, y, c=[.5, .5, .5], lw=4)

# メイン ----------------------------------
plt.figure(figsize=(4, 4))
M = 4
W = fit_gauss_func(X, T, M)
show_gauss_func(W)
plt.plot(X, T, marker='o', linestyle='None', 
         color='cornflowerblue', markeredgecolor='black')
plt.xlim(X_min, X_max)
plt.grid(True)
mse = mse_gauss_func(X, T, W)
print('W=' + str(np.round(W, 1)))
print("SD={0:.2f} cm".format(np.sqrt(mse)))
plt.show()
```

- 曲線によって誤差が3.98となり，直線の7cmより4.02cmほど小さくなった．
- フィッティングの程度を算出する平均二乗誤差を計算する関数mse_gauss_func(x, t, w)が後々やることに対して通用するので，予め関数を作っておくのは楽だと感じた．


# 5.5 オーバーフィッティングの問題

線から曲線にし，いろんな分布を表すことができるが，基底関数の数はどうやって決めるのか

```python
# リスト 5-2-(8)
plt.figure(figsize=(10, 2.5))
plt.subplots_adjust(wspace=0.3)
M = [2, 4, 7, 9]
for i in range(len(M)):
    plt.subplot(1, len(M), i + 1)
    W = fit_gauss_func(X, T, M[i])
    show_gauss_func(W)
    plt.plot(X, T, marker='o', linestyle='None', 
             color='cornflowerblue', markeredgecolor='black')
    plt.xlim(X_min, X_max)
    plt.grid(True)
    plt.ylim(130, 180)
    mse = mse_gauss_func(X, T, W)
    
    plt.title("M={0:d}, SD={1:.1f}".format(M[i], np.sqrt(mse)))
plt.show()
```

- 基底関数をそれぞれ2, 4, 7, 9にして取り組んでいる．基底関数の数を増やすと誤差の標準偏差は減る．
- ガウス基底関数は9のように乱れる．　見た感じM=7が一番フィットしている．
- この乱れる現象は　オーバーフィッティングと言われる．

```python
# リスト 5-2-(9)
plt.figure(figsize=(5, 4))
M = range(2, 10)
mse2 = np.zeros(len(M))
for i in range(len(M)):
    W = fit_gauss_func(X, T, M[i])
    mse2[i] = np.sqrt(mse_gauss_func(X, T, W))
plt.plot(M, mse2, marker='o', 
         color='cornflowerblue', markeredgecolor='black')
plt.grid(True)
plt.show()
```

- これは基底関数の数を増やした（２ー４まで）の標準偏差を表している．
- Mが増えれば増えるほど基底関数モデルは小さな動きができるようになる．
- しかしそれに伴い，標準偏差は現象するため良いが，データ点がないところで歪みが出る．
- これだと今までのデータでの誤差は小さくなっても新しいデータの予測は悪い．
- そこで一回リセット．新しいデータ予測精度を次から．

```python
# リスト 5-2-(10)
# 訓練データとテストデータ ------------------
X_test = X[:int(X_n / 4)]
T_test = T[:int(X_n / 4)]
X_train = X[int(X_n / 4):]
T_train = T[int(X_n / 4):]

# メイン ------------------------------------
plt.figure(figsize=(10, 2.5))
plt.subplots_adjust(wspace=0.3)
M = [2, 4, 7, 9]
for i in range(len(M)):
    plt.subplot(1, len(M), i + 1)
    W = fit_gauss_func(X_train, T_train, M[i])
    show_gauss_func(W)
    plt.plot(X_train, T_train, marker='o',
             linestyle='None', color='white', 
             markeredgecolor='black', label='training')
    plt.plot(X_test, T_test, marker='o', linestyle='None',
             color='cornflowerblue',
             markeredgecolor='black', label='test')
    plt.legend(loc='lower right', fontsize=10, numpoints=1)
    plt.xlim(X_min, X_max)
    plt.ylim(120, 180)
    plt.grid(True)
    mse = mse_gauss_func(X_test, T_test, W)
    plt.title("M={0:d}, SD={1:.1f}".format(M[i], np.sqrt(mse)))
plt.show()
```

- ただ，選ぶテストデータに大きく依存する．（データ数が少ない場合）
- p192の図のように，テストデータを変えるだけで，標準偏差が大きくなる．
- データ数が多ければこうならないが，今回のように4つなど少ないとこうなる．
- そこでデータ数が少なくてもばらつきを少なくする"交差検証"に取り組む．

```python
# リスト 5-2-(12)
# K 分割交差検証 -----------------------------
def kfold_gauss_func(x, t, m, k):
    n = x.shape[0]
    mse_train = np.zeros(k)
    mse_test = np.zeros(k)
    for i in range(0, k):
        x_train = x[np.fmod(range(n), k) != i]   # (A)
        t_train = t[np.fmod(range(n), k) != i]   # (A)
        x_test = x[np.fmod(range(n), k) == i]    # (A)
        t_test = t[np.fmod(range(n), k) == i]    # (A)
        wm = fit_gauss_func(x_train, t_train, m)
        mse_train[i] = mse_gauss_func(x_train, t_train, wm)
        mse_test[i] = mse_gauss_func(x_test, t_test, wm)
    return mse_train, mse_test
```

かなり改善したが，25歳のところから急激に下がっている点が次の問題．
２５歳から身長が縮んだら怖いですね（これ以上縮んでたまらん）．

そこで次の課題は25からは身長が収束していくモデルを作ること．
収束するための式はp197．グラフも詳しく書いてある.


- 関数の最小値，最大値を求める問題は最適化問題と呼ばれる．広い分野で必要とされている問題で，最適化問題を解くためのライブラリも多く開発されている．pythonですでにminimizeという関数があり，最適パラメータを出すのに今回用いる．

- minimizeの役割：最小値を求める関数とパラメータの初期値さえ与えれば，関数の微分を与えなくてもパラメータの極小解を出力してくれる．

```python
# リスト 5-2-(17)
# モデル A -----------------------------------
def model_A(x, w):
    y = w[0] - w[1] * np.exp(-w[2] * x)
    return y

# モデル A 表示 -------------------------------
def show_model_A(w):
    xb = np.linspace(X_min, X_max, 100)
    y = model_A(xb, w)
    plt.plot(xb, y, c=[.5, .5, .5], lw=4)

# モデル A の MSE ------------------------------
def mse_model_A(w, x, t):
    y = model_A(x, w)
    mse = np.mean((y - t)**2)
    return mse
```

```python
# リスト 5-2-(18)
from scipy.optimize import minimize #最適化のライブラリminimizeを呼び出す

# モデル A のパラメータ最適化 -----------------
def fit_model_A(w_init, x, t):#最適化関数で引数は左からパラメータの初期値，入力データ，目標データ
    res1 = minimize(mse_model_A, w_init, args=(x, t), method="powell")#関数内部でmse_modelを局所的に最小にしている．
    return res1.x
```

```python
# リスト 5-2-(19)
# メイン ------------------------------------
plt.figure(figsize=(4, 4))
W_init=[100, 0, 0]
W = fit_model_A(W_init, X, T)
print("w0={0:.1f}, w1={1:.1f}, w2={2:.1f}".format(W[0], W[1], W[2]))
show_model_A(W)
plt.plot(X, T, marker='o', linestyle='None', 
         color='cornflowerblue',markeredgecolor='black')
plt.xlim(X_min, X_max)
plt.grid(True)
mse = mse_model_A(W, X, T)
print("SD={0:.2f} cm".format(np.sqrt(mse)))
plt.show()
```

２５歳からの急激な下げがなくなっている．
これは$y=w0-w1*exp(-w2*x)$
w0が収束する値(今回は25歳以降)xが増加するとyはw0に近づくことになる．
