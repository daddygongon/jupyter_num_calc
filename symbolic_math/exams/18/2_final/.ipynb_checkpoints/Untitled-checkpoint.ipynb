# 1-(1) 放物線の接線の距離

````python
from sympy import *
init_printing()
a, x, t = symbols('a,x,t')
y_c = 1-x**2
plot(y_c,(x,-2,2))
````

````python
m = diff(y_c,x)
m
````

````python
y_l=collect(expand(m.subs({x: a})*(x-a)+y_c.subs({x: a})),x)
y_l
````

````python
plot(y_l.subs({a:1}),y_c,(x,-2,2))
````

点($x_0,y_0$)と直線($c_ax+c_by+c_c=0$)の距離($h$)の公式
$$
h = \frac{|c_ax_0+c_by_0+c_c|}{\sqrt{c_a^2+c_b^2}}
$$

````python
c_a=y_l.coeff(x)
c_c=y_l.coeff(x,0)
c_b=-1
````

````python
h = (c_a*0+c_b*0+c_c)/sqrt(c_a**2+c_b**2)
h
````

````python
a_2 = solve(sqrt(4*a**2+1)**2-t**2,a**2)[0]
````

````python
simplify((a_2+1)/t)
````

````python
plot(h,(a,-1,1))
````

````python
s1 = solve(diff(h,a),a)
s1
````

````python
h.subs({a:s1[1]})
````

# 1-(2) 図形の面積

2点Q($x_1,y_1$)とR($x_2,y_2$)を通る直線の方程式は
$$
y-y_1 = \frac{y_1-y_2}{x_1-x_2}(x-x_1)
$$
で求められる．

````python
from sympy import *
init_printing()
b, x, t = symbols('b,x,t')
y_c = 1-x**2
plot(y_c,(x,-2,2))
````

````python
s_c = solve(y_c,x)
s_c
````

````python
x1=s_c[0]
y1=0
x2=s_c[1]-b
y2=y_c.subs({x:x2})
````

````python
m = (y1-y2)/(x1-x2)
simplify(m)
````

````python
y_m = expand(simplify(m)*(x+1))
y_m
````

````python
plot(y_c,y_m.subs({b:0.7}),(x,-2,2))
````

````python
s1=expand(integrate(y_c-y_m,(x,-1,1-b)))
s1
````

````python
s2=expand(integrate(y_m-y_c,(x,1-b,b)))
s2
````

````python
expand(s1+s2)
````

````python
plot(s1+s2,(b,0.5,1))
````

````python
solve(diff(s1+s2,b),b)
````

# 1-(2)改

放物線$C$の方程式を$y=1-0.5x^2$として問題を解く．
放物線$C$上の2点Q($-\sqrt{2},0$)とR($\sqrt{2}-b,1-(\sqrt{2}-b)^2$)と読み替えよ．また，$S_2$を求めるときの範囲は$\sqrt{2}-b \leq x \leq b$と読み替える．

````python
from sympy import *
init_printing()
b, x, t = symbols('b,x,t')
y_c = 1-0.5*x**2
plot(y_c,(x,-2,2))
````

````python
s_c = solve(y_c,x)
s_c
````

````python
x1=s_c[0]
y1=0
x2=s_c[1]-b
y2=y_c.subs({x:x2})
````

````python
y_m = simplify((y1-y2)/(x1-x2)*(x-x1)+y1)
y_m
````

````python
s1= expand(integrate(y_c-y_m,(x,x1,x2)))
simplify(s1)
````

````python
s2=expand(integrate(y_m-y_c,(x,x2,b)))
s2
````

````python
plot(s1+s2,(b,1/2,s_c[1]))
````

````python
solve(diff(s1+s2,b),b)
````

````python
solve(simplify(diff(s1+s2,b)),b)
````

````python
solve(simplify(diff(s1+s2,b)),b)[1].evalf()
````

$$y=1-\frac{1}{2}x^2$$あるいは
``` python
y=1-Rational(1,2)*x**2
```

を使うと．答えは，
$$
\left [ - \sqrt{14} - 2 \sqrt{2}, \quad - 2 \sqrt{2} + \sqrt{14}\right ]
$$
となる．後ろ側が正解．
その場合，
``` python
solve(simplify(diff(s1+s2,b)),b)
```
としないと求められない．
