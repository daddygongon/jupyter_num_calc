# 3(a)

```python
from sympy import *
init_printing()

a, x = symbols('a x')
```

```python
C1 = Rational(1/2)*x**2 + Rational(1/2)
C1
```

```python
C2 = Rational(1/4)*x**2
C2
```

```python
S = expand(integrate(C1-C2,(x,a,a+1)))
S
```

```python
S0 = solve(diff(S,a),a)[0]
S0
```

```python
S.subs({a:S0})
```

## 3(b)

```python
solve(C1-1,x)
```

```python
solve(C2-1,x)
```
